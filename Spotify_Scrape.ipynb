```python
#Import dependencies

import pandas as pd
import os
import requests
from bs4 import BeautifulSoup as bs

# Scrape data to get list of countries available on spotify charts
url = 'https://spotifycharts.com/regional'
page = requests.get(url)
soup = bs(page.content, 'html.parser')

country_list = []

unorderedlist = soup.find_all("ul")[0].find_all('li')

for li in unorderedlist:
    country = li.get_text()
    attribute = li['data-value']
    country_list.append([country, attribute])

country_list
```

```python
# Collect the data for each country and date and write them in a list 
rows = [] 
    
#List of dates in 2020
dates_list = pd.date_range(start='2020-01-01',end='2020-01-31').strftime('%Y-%m-%d').tolist()

for country in country_list:
    for date in dates_list:
        url2 = ('https://spotifycharts.com/regional/' + country[1] +'/daily/' + date)
        page = requests.get(url2)
        soup = bs(page.content, 'html.parser')
        entries = soup.find_all("td", class_ = "chart-table-track")
        streams = soup.find_all("td", class_="chart-table-streams")
        for (i, entry, stream) in (zip(range(0,10),entries,streams)):
            song = entry.find('strong').get_text()
            artist = entry.find('span').get_text()[3:]
            play_count = stream.get_text()
            rows.append([country[0], date, song, artist, play_count, i+1])

spotify_df = pd.DataFrame(rows, columns=['Country', 'Date', 'Song_Title','Artist', 'Streams','Rank'])

spotify_df
```
